```python
## Creating a function to sketch out the distribution of certain indicators for different years 
## according to their Free Market Class
def var_dist(df, var1, year):
    plt.figure(figsize = (16,8))
    df = df[df["Year"]== year]
    sns.kdeplot(x = var1, data = df_world, hue = "Free Market Class", palette = "deep")
    plt.title("How the distribution of %s varied in %d across different %s"%(var1, year,"Free Market Class"), fontsize= 15)
    plt.xlabel("Year")
    plt.show()
```

```python
var_dist(df_world, "Human Development Index", 2019)
```

```python
var_dist(df_world, "Property Rights", 2019)
```

```python
## Function for ranking the datasets according to their continents and Free market class in a specific year
def rank_post(df, var1, var2,year):
    df = df[df["Year"]==year]
    df = df.groupby(var1, as_index = False)[var2].mean().sort_values(by = var2, ascending = False).reset_index(drop = True)
    plt.figure(figsize = (16,8))
    sns.barplot(x = var1, y = var2, data = df)
    plt.xlabel(var1, fontsize = 12)
    plt.ylabel(var2, fontsize = 12)
    plt.title("Displaying the average %s across %s"%(var2, var1), fontsize = 14)
    plt.show()
```

```python
rank_post(df_world, "Free Market Class", "Income", 2019)
```

```python
rank_post(df_world, "Continent", "Income", 2019)
```
